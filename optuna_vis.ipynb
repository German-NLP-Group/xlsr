```python
import optuna
import pandas as pd

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_colwidth', None)
```

```python
# the target is 0.8662709406419864
# base model: # de 0.8078971446757308 en 0.8349525108990091 cross 0.798345566174531

study_name = "xlsr_de_en_cross_stsb_08"
#study_name = "xlsr_de_en_cross_stsb_08_do_01"  # both dropouts
#study_name = "xlsr_de_en_cross_stsb_08_do_02"  # hidden_dropout_prob only
#study_name = "xlsr_de_en_cross_stsb_08_xnli_02"  # add xnli

storage='sqlite:///optuna.db'
```

```python
study = optuna.create_study(study_name=study_name, 
                            storage=storage,
                            load_if_exists=True, direction='maximize')

df = study.trials_dataframe()

df_gt_0 = df[df['value'] > 0.0]
print('trials with value > 0.0:', len(df_gt_0))

df.sort_values('value', ascending=False, na_position='first')
```

```python
best_trial = study.best_trial
print(best_trial)
print('##############')
print(best_trial.params)
```

```python
trials = [t for t in study.get_trials() if t.value is not None]
trials = sorted(trials, key=lambda trial: trial.value, reverse=True)
trials = trials[:5]
for t in trials:
    print(t.params)
```

```python
fig = optuna.visualization.plot_slice(study)
#fig.write_image(f"./plots/{study_name}.png")  # save to image
fig.write_html(f"./plots/{study_name}_slice.html")  # save to html
```
