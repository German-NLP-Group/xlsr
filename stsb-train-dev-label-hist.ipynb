# Histogram of STSb train and dev Labals 

```python
# imports
import csv
import matplotlib.pyplot as plt
```

```python
def load_data_file_similarity_scores(data_file):
    similarity_scores = []
    sentence_len = []
    with open(data_file, newline="", encoding="utf-8") as csvfile:
        csv_dict_reader = csv.DictReader(
            csvfile,
            dialect='excel',
            fieldnames=["sentence1", "sentence2", "similarity_score"],
        )
        for row in csv_dict_reader:
            # load similarity_score
            ss = float(row['similarity_score'])
            assert ss >= 0.0
            assert ss <= 5.0
            similarity_scores.append(ss)

            # load sentence length
            assert "sentence1" in row
            assert isinstance(row["sentence1"], str)
            assert len(row["sentence1"].strip()) > 0
            assert len(row["sentence1"].strip()) == len(row["sentence1"])
            sentence_len.append(len(row["sentence1"].strip()))
            assert "sentence2" in row
            assert isinstance(row["sentence2"], str)
            assert len(row["sentence2"].strip()) > 0
            assert len(row["sentence2"].strip()) == len(row["sentence2"])
            sentence_len.append(len(row["sentence2"].strip()))
    return similarity_scores, sentence_len
```

```python
scores, sentece_len = load_data_file_similarity_scores("./data/stsb-en-train.csv")
s2, sl2 = load_data_file_similarity_scores("./data/stsb-en-dev.csv")
scores.extend(s2)
sentece_len.extend(sl2)
```

```python
_ = plt.hist(scores, bins=5)
```

```python
_ = plt.hist(sentece_len, bins=20)
```

```python
_ = plt.boxplot(sentece_len)
```
